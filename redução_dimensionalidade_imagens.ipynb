```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

def reduzir_dimensoes(imagem_path):
    # Abre a imagem
    imagem = Image.open(imagem_path)

    # Converte para escala de cinza
    imagem_cinza = imagem.convert('L')

    # Converte para matriz numpy
    imagem_array = np.array(imagem)
    imagem_cinza_array = np.array(imagem_cinza)

    # Reduz para preto e branco
    imagem_pb = np.where(imagem_cinza_array < 128, 0, 255).astype(np.uint8)

    # Cria imagens PIL a partir dos arrays
    imagem_reduzida_cinza = Image.fromarray(imagem_cinza_array)
    imagem_reduzida_pb = Image.fromarray(imagem_pb)

    # Exibe as imagens
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    axes[0].imshow(imagem_array)
    axes[0].set_title('Original')
    axes[0].axis('off')
    axes[1].imshow(imagem_reduzida_cinza, cmap='gray')
    axes[1].set_title('Escala de Cinza')
    axes[1].axis('off')
    axes[2].imshow(imagem_reduzida_pb, cmap='gray')
    axes[2].set_title('Preto e Branco')
    axes[2].axis('off')
    plt.show()

# Exemplo de uso
reduzir_dimensoes('/content/truck.jpg')
```
